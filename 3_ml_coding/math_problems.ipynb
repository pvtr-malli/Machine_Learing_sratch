# Q1: Calculate Cosine Similarity Between Vectors
formula:
$$
similarity = \frac{a.b}{||a|| *  ||b||}
$$
* (|A|), (|B|) = Euclidean norms (l2 norm or magnitudes) of the vectors


```python
import numpy as np

def cosine_similarity(v1, v2):
	# Implement your code here
    a = np.dot(v1, v2)
    v1_distance = np.sqrt(np.sum(np.square(v1))) # finding the l2 norms
    v2_distance = np.sqrt(np.sum(np.square(v2)))
    return round(a / (v1_distance * v2_distance), 3)


v1 = np.array([1, 2, 3])
v2 = np.array([2, 4, 6])
print(cosine_similarity(v1, v2))
```
