# 9 | Stacking & Concatenation

```python
np.concatenate([a, b], axis=0)
np.vstack([a, b])
np.hstack([a, b])
```

```python
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])
```

```python
np.concatenate([a, b], axis=0) # row wise stacking vertical stacking
```

```python
np.concatenate([a, b], axis=1) # column stacking horizondal stacking
```

```python
np.vstack([a, b]) # same as axis = 0
```
